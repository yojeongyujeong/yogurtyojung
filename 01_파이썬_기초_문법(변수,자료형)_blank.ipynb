### (4) 수정/추가/삭제

#### 수정

````python
# 0에서 7까지의 범위를 리스트로 출력
num_list = list(range(8))
print(num_list)

# 1번째 값을 '일'로 바꾸기
print(num_list[1])
num_list[1] = "일"
print(num_list)

# 5번째 값을 '오'로 바꾸기
num_list[1] = 1
num_list[5] = "오"
print(num_list)


# 2번째에서 3번째까지의 데이터를 영어로 바꾸기
num_list[5] = 5
num_list[2:4] = ["two", "three"]
print(num_list)

# 3번째에서 끝까지의 데이터를 "하하하"로 바꾸기
num_list[2] = 2
num_list[3] = 3
num_list[3:] = "하하하히" # 문자열은 이렇게 쓰면 열로 인식
print(num_list)

# 처음부터 3번째까지의 데이터를 10으로 바꾸기
num_list[:4] = [10, 10, 10, 10]
num_list[3:] = [4, 5, 6, 7]
print(num_list)
````

### (3) 연산

<img src='https://www.dropbox.com/scl/fi/o9jjkizgbclwemi9wot4h/.png?rlkey=35o0zu24jc99ggiv9ex3pukqf&st=f69cko7q&dl=1' width=30%>

````python
in_likesport = {"농구", "축구", "야구", "배구", "MMA"}
chan_likesport = {"배구", "수영", "양궁", "축구", "클라이밍"}

# 합집합
set_union = in_likesport | chan_likesport
print(set_union)
set_union2 = in_likesport.union(chan_likesport)
print(set_union2)

# 교집합
set_union = in_likesport & chan_likesport
print(set_union)
set_union2 = in_likesport.intersection(chan_likesport)
print(set_union2)

# 차집합 (위 - 아래)
set_union = in_likesport - chan_likesport
print(set_union)
set_union2 = in_likesport.difference(chan_likesport)
print(set_union2)

# 차집합 (아래 - 위)
set_union = chan_likesport - in_likesport
print(set_union)
set_union2 = chan_likesport.difference(in_likesport)
print(set_union2)
````

````python
# * append / extend 차이 알기
# 인스타그램 태그 수집기 
# post1 : #헬스, #단백질
# post2 : #닭가슴살, #다이어트, #단백질
# post3 : #요요, #헬스, #다이어트
# post4 : #닭가슴살, #단백질

# 문제 1. 인스타그램 포스트별로 데이터를 수집하고 싶다
# 문제 2. 모든 해시태그들을 tags라는 변수에 담고 싶다
# 문제 3. 해시태그 카테고리가 몇 개인지 알고 싶다


# 문제 1. 인스타그램 포스트별로 데이터를 수집하고 싶다

post1 = ["#헬스", "#단백질"]
post2 = ["#닭가슴살", "#다이어트", "#단백질"]
post3 = ["#요요", "#헬스", "#다이어트"]
post4 = ["#닭가슴살", "#단백질"]

posts = [post1, post2, post3, post4]
tags = []
for post in posts:
    tags.append(post)

print(tags)

# 문제 2. 모든 해시태그들을 tags라는 변수에 담고 싶다
tags = []
for post in posts:
    tags.extend(post)

print(tags)

# 문제 3. 해시태그 카테고리가 몇 개인지 알고 싶다
result = set(tags)
print(result)
````
