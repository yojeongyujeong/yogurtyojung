### 2) 순서형 데이터

````python
# 만족도에 대한 막대 그래프를 그려주세요.
import seaborn as sns 
import matplotlib.pyplot as plt
import koreanize_matplotlib
import platform

plt.figure(figsize=(3, 3))
sns.countplot(data=variation_data, x="만족도")
plt.title("만족도에 대한 막대그래프")

plt.figure(figsize=(3, 3))
gender_data = variation_data["만족도"].value_counts()
gender_data.plot(kind="bar")
plt.title("만족도에 대한 막대그래프")
````

````python
# 성별에 따른 만족도에 대해 시각화 하고 싶어요.
# 만족도에 대한 막대 그래프를 그려주세요.
import seaborn as sns 
import matplotlib.pyplot as plt
import koreanize_matplotlib
import platform

plt.figure(figsize=(3, 3))
sns.countplot(data=variation_data, x="만족도", order=["하", "중", "상"], hue="성별")
plt.title("만족도에 대한 막대그래프")
plt.show()
````

### 3) 연속형 데이터

````python
# "월 수입(만원)"열에 대한 히스토그램
plt.figure(figsize=(3, 3))
plt.hist(variation_data["월 수입(만원)"], bins=20, edgecolor="black")
plt.title("월 수입(만원) 히스토그램")
plt.show()

# "월 수입(만원)"열에 대한 히스토그램
plt.figure(figsize=(3, 3))
plt.hist(variation_data["월 수입(만원)"], bins=50, edgecolor="black")
plt.title("월 수입(만원) 히스토그램")
plt.show()

# "월 수입(만원)"열에 대한 히스토그램
plt.figure(figsize=(3, 3))
plt.hist(variation_data["월 수입(만원)"], bins=80, edgecolor="black")
plt.title("월 수입(만원) 히스토그램")
plt.show()
````

````python
# "월 수입(만원)"열에 대한 boxplot
plt.figure(figsize=(3, 3))
plt.boxplot(x=variation_data["월 수입(만원)"])
plt.title("월 수입(만원)에 대한 BoxPlot")
plt.xlabel("월 수입(만원)")
plt.show()
````

### 4) 이산형 데이터

````python
# 방문 페이지 수
## 구글링 : ax를 이용한 시각화 도전

fig, ax = plt.subplots()
sns.countplot(data=variation_data, x="방문 페이지 수", ax=ax)
ax.set_title("방문 페이지 수에 대한 막대그래프")
plt.show()
````

### 5) 시계열 데이터

#### 날짜에 대한 데이터 타입은 `datetime`이다.

````python
variation_data.info() # 결측치 없다!
variation_data["날짜"] = pd.to_datetime(variation_data["날짜"]) # 날짜 데이터로 지정해서 가지고 오는법(1)
variation_data.info()
````

````python
date_data = pd.read_csv("data/sample_data.csv", parse_dates=["날짜"]) # 날짜 데이터로 지정해서 가지고 오는법(2)
date_data.info()
````

````python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(variation_data["날짜"], variation_data["평균 체류 시간(분)"],color="skyblue", marker="o")
ax.set_title("날짜별 평균 체류 시간 변화 그래프")
ax.set_xlabel("날짜")
ax.set_ylabel("평균 체류 시간(분)")
plt.show()
````

````python
fig, ax = plt.subplots(1, 3, figsize=(10,3))
# 첫번째 그래프: 자녀 수에 대한 막대 그래프
sns.countplot(data=variation_data, x="자녀 수", ax=ax[0], color="pink", edgecolor="black")
ax[0].set_title("자녀 수에 대한 막대그래프")
# 두번째 그래프: 평균 체류 시간(분)에 대한 히스토그램
ax[1].hist(variation_data["평균 체류 시간(분)"], color="pink", edgecolor="black")
ax[1].set_title("평균 체류 시간(분)에 대한 히스토그램")
# 세번째 그래프: 월 수입(만원)에 대한 상자그림
ax[2].boxplot(x=variation_data["월 수입(만원)"])
ax[2].set_title("월 수입(만원)에 대한 Box Plot")
ax[2].set_xlabel("월 수입(만원)")
````
