## 1) 연산자

문제. 나이가 10세 이상이고, 키가 120cm 이상일 때만 놀이기구 탑승

* 탑승 가능하면 "통과" 아니면 "탑승 불가"라고 출력하기

````python
# 나이가 10세 이상이고, 키가 120cm 이상일 때만 놀이기구 탑승
# 탑승 가능하면 "통과" 아니면 "탑승 불가"라고 출력하기

age = int(input("친구 몇 살이에요?"))
height = float(input("키를 입력해볼까요, 친구?(cm)"))

if age >= 10 and height >= 120.0 :
    print("통과")
else:
    print("오우, 저런.. 다음기회에 다시 도전하세요ㅜ")
````

문제. 10만원 이상 구매하거나 "Gold" 등급이면 10% 할인

* 할인이 가능하면 "10% 할인 가능" 아니면 "10% 할인 불가" 출력

````python
# 10만원 이상 구매하거나 "Gold" 등급이면 10% 할인
# 할인이 가능하면 "10% 할인 가능" 아니면 "10% 할인 불가" 출력

purchase_price = int(input("그대의 구매금액은 얼마입니까?(원단위)"))
customer_grade = str(input("그대의 등급을 알려줘요..(gold, silver, copper 중 1개를 입력하세요)"))

if purchase_price >= 100000 or customer_grade == "gold":
    print("10% 할인 가능")
else:
    print("10% 할인 불가")
````

## 2) 활용

````python
# 지갑 안에 운전면허증(license)이 있다면 통과이고 없으면 벌금이다. 
# 만약 벌금을 납부하게 되면 지갑에 카드(credit card)가 있으면 카드 결제이고 아니면 현금 납부이다. 

# 벌금을 납부하게 된다면 "벌금 대상자"을 출력하고 아니면 "통과"를 출력한다
# 그리고 벌금을 납부하게 된다면 "카드납부" 아니면 "현금납부"
# 이 사람은 어떻게 해야 하는가?
wallet = ["credit card", "trash"]

if "license" in wallet:
    print("통과")
elif "credit card" in wallet:
    print("벌금 대상자이며, 카드 납부")
else:
    print("벌금 대상자이며, 현금 납부")
````

* GPA가 3.0 이상이면 30% 장학금을 받습니다.

### (3) break/continue

#### break

````python
# 1부터 10까지 숫자를 하나씩 출력 만약 7이 나오면 반복문 중지
count = 0
while True:        # while True: -> 무한루프
    count += 1
    print(count)

    if count >= 7:
        break
````

````python
# 1부터 10까지 숫자를 하나씩 출력 만약 7이 나오면 반복문 중지
for i in range(1, 11):
    print(i)
    if i == 7:
        break
````

#### continue

````python
# 1부터 10까지 숫자 중에서 5만 빼고 출력하기
count = 0
while count < 10:
    count += 1

    if count == 5:
        continue
    
    print(count)
````

## 4) List Comprehension

````python
# 1~10까지의 숫자 리스트를 만들기
num_list = [i for i in  range(1, 11)]
print(num_list)
````

````python
# 1~10까지의 숫자 제곱 리스트를 만들기
num_list = [i**2 for i in  range(1, 11)]
print(num_list)
````

````python
# 1~10까지의 짝수/홀수 리스트를 만들고 싶다.
num_list = [i for i in  range(1, 11) if i % 2 == 0]
print(num_list)
num_list = [i for i in  range(1, 11) if i % 2 == 1]
print(num_list)
````

````python
# 1~10까지의 짝수/홀수 리스트("짝수", "홀수"쓰여진)를 만들고 싶다.
num_list = ["짝수" if i % 2 == 0 else "홀수" for i in  range(1, 11)]
print(num_list)
````

## 5) Dictionary Comprehension

````python
wishlist = {"향수": 80000, "미니건조기": 299000, "키보드": 20000, "이동식메모리": 30000}

# 20만원 이하인 상품들만 딕셔너리로 만들고 싶어요
wishlist = {"향수": 80000, "미니건조기": 299000, "키보드": 20000, "이동식메모리": 30000}
purch_list = {limit_approve_list: limit for limit_approve_list, limit in wishlist.items() if limit <= 200000} 
print(purch_list)
````
