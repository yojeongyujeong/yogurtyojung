**넘파이는 C언어로 만들어져 속도도 빠르고, 연산도 쉽게 할 수 있습니다.**

```python
from PIL import Image
from IPython.display import display

image = Image.open("mycuteimage/꼬양이.jpg")
display(image)
```

```python
import numpy as np 
image_arr = np.array(image)
image_arr
```

```python
print(image_arr.shape) #(a, b, c) -> c가 rgb라서 3개!
print(image_arr.ndim)
```

```python
image_arr2 = 255 - image_arr         # 최대 255
image2 = Image.fromarray(image_arr2)

display(image2)
```
