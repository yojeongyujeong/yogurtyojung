```python
# x, y를 넣으면 덧셈해주는 함수

def only_plus():
    x = int(input("숫자 입력해주세요."))
    y = int(input("숫자 입력해주세요."))
    sum = x + y
    return sum

only_plus()
```

```python
# x, y를 넣으면 덧셈해주는 함수

def only_plus2(x:int, y:int):
    return x + y

result = only_plus2(55, 9)
print(result)
```

## 3) 매개변수

```python
# 이름과 나이를 입력하면 "김유정님의 나이는 27세입니다." 라고 출력해주는 함수

def per_info(name, age, gender="여성", details=None):
    print(f"{name}님의 나이는 {age}세 {gender}입니다.")
    print(f"상세사항: {details}")

per_info("김유정", 27, "여성", "인천 사람입니다.")
```

```python
# *args -> 튜플형, **kwargs -> 딕셔너리
# 시험 성적의 평균을 반환하는 함수

def avg_score(*args):
    count = len(args)
    avg = sum(args) / count

    return avg

avg_score(100, 67, 87, 91, 35)

# **kwargs -> 딕셔너리
# 학생의 시험점수를 한눈에 보이게 출력해주는 함수
# name : 김유정
# math : 100
# ...

def prt_score(**kwargs):
    print(kwargs)
    for key, value in kwargs.items():
        print(f"{key}:{value}")

prt_score(name="김유정", math=100, science=100, tech=100)
```

## 4) lambda 함수

```python
# x, y를 입력했을 때 덧셈해주는 함수
def only_plus2(x:int, y:int):
    return x + y

print(only_plus2(55, 9))
```

```python
only_plus2 = lambda x, y : x + y
print(only_plus2(15, 10))
```
